```python
P14retina_1_PCA.plot()
```

```python
%time P14retina_1_tSNE = fcc_tSNE(P14retina_1_PCA.results, perplexity=30) # perform tSNE on 100-component PCA results
```

```python
P14retina_1_tSNE.clu.draw_decision_graph() # look at potential cluster centers
```

```python
# assign cluster centers and visualize
P14retina_1_tSNE.clu.assign(100,5)
P14retina_1_tSNE.clu.draw_decision_graph(min_delta=P14retina_1_tSNE.clu.min_delta, min_density=P14retina_1_tSNE.clu.min_density)
```

```python
# show clustering and silhouette score
P14retina_1_tSNE.plot_clusters()
```

```python
P14retina_1_tSNE.silhouette_score
```

Let's try to mash the replicates together and analyze

```python
print(P14retina_1.data.shape)
print(P14retina_3.data.shape)
print(P14retina_5.data.shape)
print(P14retina_7.data.shape)
```

```python
P14retina_1_3 = P14retina_1.data.append(P14retina_3.data)
P14retina_5_7 = P14retina_5.data.append(P14retina_7.data)
```

```python
P14retina = P14retina_1_3.append(P14retina_5_7)
P14retina.shape
```

```python
P14retina_combined = RNA_counts(data=P14retina.fillna(value=0))
```

```python
P14retina_combined.random_subset(n_cells=400)
```

```python
P14retina_combined_PCA = fcc_PCA(P14retina_combined.arcsinh_norm(), n_components=100)
```

```python
P14retina_combined_PCA.plot()
```

```python
P14retina_combined_tSNE = fcc_tSNE(P14retina_combined_PCA.results, perplexity=30)
```

```python
P14retina_combined_tSNE.plot()
```

```python
P14retina_combined_tSNE.clu.draw_decision_graph()
```

```python
P14retina_combined_tSNE.clu.assign(5,3)
P14retina_combined_tSNE.clu.draw_decision_graph(min_delta=P14retina_combined_tSNE.clu.min_delta, min_density=P14retina_combined_tSNE.clu.min_density)
```

```python
P14retina_combined_tSNE.plot_clusters()
```

```python
### from fcc_utils import compare_euclid
%time mantel_stats, EMD, KLD = compare_euclid(pre=P14retina_combined_PCA.distance_matrix(), post=P14retina_combined_tSNE.distance_matrix())
```

```python
import nvr
```

```python
hqGenes=nvr.parseNoise(P14retina_combined.counts)
dHq=nvr.mkIndexedArr(P14retina_combined.counts,hqGenes)
dHq.shape
```

```python
dataHq=nvr.pwArcsinh(dHq,1000)
```

```python
P14retina_combined_selected_genes=nvr.select_genes(dataHq)
```

```python
P14retina_combined_selected_genes.shape
```

```python
dataHq[:,P14retina_combined_selected_genes].shape
```

```python
P14retina_combined_nvr = RNA_counts(dataHq[:,P14retina_combined_selected_genes], labels=[None,None])
P14retina_combined_nvr.counts.shape
```

```python
%time P14retina_combined_nvr_PCA = fcc_PCA(P14retina_combined_nvr.arcsinh_norm(), n_components=100)
```
